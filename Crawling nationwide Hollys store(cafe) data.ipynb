```python
#메장명에 대한 Xpath
#  //*[@id="contents"]/div[2]/fieldset/fieldset/div[1]/table/tbody/tr[1]/td[2]
import requests
import bs4

page_no = 1
hollys_url = f"https://www.hollys.co.kr/store/korea/korStore2.do?pageNo={page_no}&sido=&gugun=&store="

source = requests.get(hollys_url).text
source = bs4.BeautifulSoup(source)

regions = source.find_all('td', class_='noline center_t')
regions
```

```python
regions[0].text
```

```python
region_list = []

for region in regions:
    region_list.append(region.text)
    
region_list
```

```python
names = source.find_all('td', class_='center_t')
name_list = []     # [0:10:2] 0부터 10-1까지 2칸씩 점프
for name in names[1:60:6]:  #6번째부터 전화번호 시작하니까, 6부터 60번까지 6의 배수로 돌리기
    name_list.append(name.text)
name_list
```

```python
phones = source.find_all('td', class_='center_t')
phone_list = []     # [0:10:2] 0부터 10-1까지 2칸씩 점프
for phone in phones[5:60:6]:  #6번째부터 전화번호 시작하니까, 6부터 60번까지 6의 배수로 돌리기
    phone_list.append(phone.text)
phone_list
```

```python
region_list = []
name_list = []
phone_list = [] 

for page_no in range(1, 53):
    hollys_url = f"https://www.hollys.co.kr/store/korea/korStore2.do?pageNo={page_no}&sido=&gugun=&store="

    source = requests.get(hollys_url).text
    source = bs4.BeautifulSoup(source)

    regions = source.find_all('td', class_='noline center_t')
    
    for region in regions:
        region_list.append(region.text)
        
    names = source.find_all('td', class_='center_t')
    
    for name in names[1:60:6]:
        name_list.append(name.text)
    
    phones = source.find_all('td', class_='center_t')
       
    for phone in phones[5:60:6]:
        phone_list.append(phone.text)
```

```python
len(region_list), len(phone_list)
```

```python
import pandas as pd

hollys_tbl = pd.DataFrame({"region": region_list, "center_name": name_list, "phone number": phone_list})
hollys_tbl
```

```python
hollys_tbl.to_excel("hollys.xlsx", index=False)
```
